## Analyse audio file using librosa library

<small>dataset url: https://urbansounddataset.weebly.com/urbansound8k.html</small>

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa
import IPython.display as ipd
import pandas as pd
```

<small>Librosa loads the audio file and resample to a sample rate of 22050 Hz. The "mono=True" argument ensures that stereo audio files are loaded as mono. The "sr=22050" argument specifies the target sample rate. It also gives a normalized dataset</small>

```python
filename='UrbanSound8K/car_horn.wav'

# Load the audio file
data, sample_rate = librosa.load(filename, sr=22050, mono=True)

# Calculate time array corresponding to each sample
time = np.arange(0, len(data)) / sample_rate

# Plot the waveform using matplotlib
plt.figure(figsize=(14, 5))
plt.plot(time, data)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Waveform')
plt.show()

# Display the audio
ipd.Audio(filename)
```
